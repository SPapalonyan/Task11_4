```
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error


def car_price_regression(filename: str):
    # 1. Загрузка данных
    df = pd.read_csv(filename)

    # 2. Выбор признаков и целевой переменной
    features = ['Year', 'Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'Popularity']
    target = 'MSRP'

    X = df[features]
    y = df[target]

    # 3. Обработка пропущенных значений (замена на среднее)
    X = X.fillna(X.mean())

    # 4. Разделение на тренировочную и тестовую выборки (75/25)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

    # 5. Обучение модели линейной регрессии
    model = LinearRegression()
    model.fit(X_train, y_train)

    # 6. Предсказания на тестовой выборке
    y_pred = model.predict(X_test)

    # 7. Вычисление RMSE для линейной регрессии
    rmse_lr = root_mean_squared_error(y_test, y_pred)

    # 8. "Глупая модель" — предсказывает среднее значение из тренировочной выборки
    y_mean = y_train.mean()
    y_pred_dummy = [y_mean] * len(y_test)

    # 9. Вычисление RMSE для "глупой модели"
    rmse_dummy = root_mean_squared_error(y_test, y_pred_dummy)

    # 10. Вывод результатов
    print(f"RMSE линейной регрессии: {rmse_lr:.2f}")
    print(f"RMSE глупой модели:      {rmse_dummy:.2f}")
    print(f"Разница (глупая - LR):   {rmse_dummy - rmse_lr:.2f}")

    return rmse_lr, rmse_dummy


if __name__ == "__main__":
    url = "https://raw.githubusercontent.com/SPapalonyan/Task11_4/main/data.csv"
    car_price_regression(url)
```
